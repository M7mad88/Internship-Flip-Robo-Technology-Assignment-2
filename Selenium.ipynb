```python
# 6) 

def flip_sneakers(url):
    
    # Let's load the drivers and URL
    browser= webdriver.Chrome()
    browser.get(url)
    
    #Let's navigate to search bar
    browser.find_element(By.XPATH,'//button[@class="_2KpZ6l _2doB4z"]').click()
    time.sleep(3)
    
    #Let's search for sneaker product and then click search button
    browser.find_element(By.XPATH,'//input[@title="Search for products, brands and more"]').send_keys('sneakers')
    browser.find_element(By.XPATH,'//button[@type="submit"]').click()
    time.sleep(3)
    
    # Let's create empty list for storing the data
    sneaker_brand_name=[]
    sneaker_description=[]
    sneaker_price=[]
    sneaker_discount=[]
    
    # Let's scrape the data
    j=0
    while j<3:
        time.sleep(3)
        brands=browser.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
        for brand in brands:
            sneaker_brand_name.append(brand.text)
        
        descriptions=browser.find_elements(By.XPATH,'//a[contains(@class,"IRpwTa")]')
        for description in descriptions:
            sneaker_description.append(description.text)
        
        prices=browser.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
        for price in prices:
            sneaker_price.append(price.text)
        
        discounts=browser.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
        for discount in discounts:
            sneaker_discount.append(discount.text)
        if j==0:
            browser.find_element(By.XPATH,'//a[@class="_1LKTO3"]').click()
        elif j==1:
            browser.find_element(By.XPATH,'//nav[@class="yFHi8N"]/a[12]').click()
        else:
            pass
        j+=1
        
    # Let's create a DataFrame for our data   
    sneakers_flipkart=pd.DataFrame({})
    sneakers_flipkart['Brand Name']=sneaker_brand_name[:100]
    sneakers_flipkart['Product Descriptions']=sneaker_description[:100]
    sneakers_flipkart['Price']=sneaker_price[:100]
    sneakers_flipkart['Discount %']=sneaker_discount[:100]
    return sneakers_flipkart
```

```python
# Let's call the Calling Function

flip_sneakers('https://www.flipkart.com/')
```

```python
# 7)

#Let's call the Function Definition

def laptop(url):
    
    # Let's load the drivers and URL
    browser= webdriver.Chrome()
    browser.get(url)
    
    
    # Let's create empty list for storing the data
    item_title=[]
    price=[]
    rating=[]
    
    #Let's search for laptop product and then click search button
    browser.find_element(By.XPATH,'//input[@id="twotabsearchtextbox"]').send_keys('Laptop')
    browser.find_element(By.XPATH,'//input[@type="submit"]').click()
        
    #Let's apply filter for "Intel Core i7"
    browser.find_element(By.XPATH,'//*[@id="p_n_feature_thirteen_browse-bin/12598163031"]/span/a',).click()
    time.sleep(5)
        
    # Let's scrape the data
    titles=browser.find_elements(By.XPATH,'//span[@class="a-size-medium a-color-base a-text-normal"]')
    for i in titles:
        item_title.append(i.text)
    prices=browser.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
    for i in prices:
        price.append(i.text)
    ratings=browser.find_elements(By.XPATH,'//div[@class="a-row a-size-small"]/span[1]')   
    for i in ratings:
        rating.append(i.get_attribute('aria-label'))
        
    # Let's create a DataFrame for our data
    amazon_laptops=pd.DataFrame({})
    amazon_laptops['Title']=item_title[:10]
    amazon_laptops['Price']=price[:10]
    amazon_laptops['Rating']=rating[:10]
    return amazon_laptops
```

```python
# Let's call the Calling Function

laptop('https://www.amazon.in/')
```

```python
# 8)

def Quotes1000(url):
    
    # Let's load the drivers and URL
    browser= webdriver.Chrome()
    browser.get(url)
    
    # Let's create empty list for storing the data
    Quotes=[]
    Authors=[]
    quotes_types=[]
    
    
    #Let's click on top quotes button
    browser.find_element(By.XPATH,'//*[@id="menu"]/div/div[3]/ul/li[5]/a').click()
    
   # Let's scrape the data
    j=0
    while j<10:
        time.sleep(3)
        quotes=browser.find_elements(By.XPATH,'//a[@class="title"]')
        for quote in quotes:
            Quotes.append(quote.text)
        
        authors=browser.find_elements(By.XPATH,'//div[@class="author"]')
        for author in authors:
            Authors.append(author.text)
        
        qt=browser.find_elements(By.XPATH,'//div[@class="tags"]')
        for q in qt:
            quotes_types.append(q.text)
        time.sleep(5)
    
        if j<9:
            browser.find_element(By.XPATH,'//*[@id="content"]/div/div[1]/div/div[4]/li[12]/a').click()
        j+=1
        
        
    # Let's create a DataFrame for our data
    df=pd.DataFrame({})
    df['Quotes']=Quotes
    df['Authors']=Authors
    df['Type of Quotes']=quotes_types
    return df    
```

```python
# Let's call the Calling Function

Quotes1000('https://www.azquotes.com/')
```

```python
# 9)

def j_josh(url):
    
    # Let's load the drivers and URL
    browser= webdriver.Chrome()
    browser.get(url)
    
    # Let's create empty list for storing the data
    Name=[]
    Born_Dead=[]
    Term_of_office=[]
    Remarks=[]
    
    
    #Let's click on GK button
    browser.get('https://www.jagranjosh.com/')
    time.sleep(3)
    browser.find_element(By.XPATH,'//*[@id="close"]').click()
    browser.find_element(By.XPATH,'//*[@id="1540978020504"]/div[1]/header/div[3]/ul/li[9]/a').click()
    time.sleep(3)
    browser.find_element(By.XPATH,'//*[@id="popluarGK"]/ul/li[2]/a').click()
    
   # Let's scrape the data
    quotes=browser.find_elements(By.XPATH,'//tr[@style]')
    for i in quotes[1:]:
        list_=i.text.split('\n')[1:]
        Name.append(list_[0])
        Born_Dead.append(list_[1])
        Term_of_office.append(' / '.join(list_[2:4]))
        Remarks.append(list_[-1])
        
        
    # Let's create a DataFrame for our data
    df=pd.DataFrame({})
    df['Names']=Name
    df['Born/Dead']=Born_Dead
    df['Term of Office']=Term_of_office
    df['Remarks']=Remarks
    return df    
```

```python
j_josh('https://www.jagranjosh.com/')
```

```python
# 10)

def motor1(url):
    
    # Let's load the drivers and URL
    browser= webdriver.Chrome()
    browser.get(url)
    # Let's create empty list for storing the data
    Car_name=[]
    Description=[]
    Price=[]
    
    
    
    #Let's click on dropdown button then features list.
    browser.find_element(By.CLASS_NAME,'m1-hamburger-button').click()
    time.sleep(5)
    browser.find_element(By.XPATH,'/html/body/div[4]/div[1]/div[3]/ul/li[5]/button').click()
    browser.find_element(By.XPATH,'/html/body/div[4]/div[1]/div[3]/ul/li[6]/ul/li[1]/a').click()
    time.sleep(2)
    browser.find_element(By.XPATH,'//*[@id="page_index_articles_browse"]/div[9]/div[1]/div[1]/div/div/div[2]/div/div[1]/h3/a').click()
    
   # Let's scrape the data
    names=browser.find_elements(By.XPATH,'//h3[@class="subheader"]')
    for i in names[:-1]:
        Car_name.append(i.text)
    
    desc=browser.find_elements(By.XPATH,'//*[@id="article_box"]/div[1]/div[2]/div[1]/p')
    for i in desc[4:103:2]:
        Description.append(i.text)
            
    prices=browser.find_elements(By.XPATH,'//strong')
    for i in prices[:-1]:
        if 'Price:' in i.text:
            Price.append(i.text[7:])
        else:
            Price.append(i.text)
        
        
    # Let's create a DataFrame for our data
    df=pd.DataFrame({})
    df['Names']=Car_name
    df['Description']=Description
    df['Price']=Price
    return df    
```

```python
motor1('https://www.motor1.com/')
```
